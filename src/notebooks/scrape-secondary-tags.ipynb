```python
from bs4 import BeautifulSoup
import requests
import yaml
```

```python
TAGS_URL = "https://genius.com/Genius-tags-music-genres-countries-languages-annotated"
```

```python
html_doc = requests.get(TAGS_URL).text
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
links = [link.get('href') for link in soup.find_all('a')]
links
```

```python
links = [link for link in links if "https://genius.com/tags/" in link]
links 
```

```python
genres = [link[24:] for link in links]
genres 
```

```python
secondary_genres = genres[5:]
secondary_genres
```

```python
with open("../../data/secondary_genres.yaml", "w") as f:
    yaml.safe_dump(secondary_genres, f)
```
